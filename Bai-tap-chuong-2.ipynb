## Bài 2.1 – XML công ty

```
xml_content = """<?xml version="1.0" encoding="UTF-8"?>
<company>
    <name>Công ty TNHH Đại Việt</name>
    <director>
        <name>Nguyễn Văn A</name>
    </director>
    <address>abc Giải Phóng - Hà Nội</address>
    <phone>0123456789</phone>
    <tax_code>0101234567</tax_code>
</company>
"""
print(xml_content)
```

## Bài 2.4 & 2.5 – Đọc XML bằng minidom

```
from xml.dom import minidom

doc = minidom.parseString(xml_content)
company = doc.documentElement

print("Tên công ty:", company.getElementsByTagName("name")[0].firstChild.data)
```

## Bài 2.7 – JSON → Python

```
import json

json_str = '{"name":"An","age":20,"city":"Ha Noi"}'
data = json.loads(json_str)
print(data)
```

## Bài 2.8 – Python → JSON

```
json_out = json.dumps(data, ensure_ascii=False, indent=4)
print(json_out)
```

## Bài 2.12 – Đọc API Post

```
import requests

url = "https://jsonplaceholder.typicode.com/posts"
posts = requests.get(url).json()

print("Tổng số bài post:", len(posts))
print(posts[0])
```
