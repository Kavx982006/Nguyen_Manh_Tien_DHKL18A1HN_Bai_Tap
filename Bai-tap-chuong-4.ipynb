## Bài 4.1

```
mylist = list('abcdefghijklmnopqrstuvwxyz')
myarr = np.arange(26)
mydict = dict(zip(mylist, myarr))

ser_list = pd.Series(mylist)
ser_arr = pd.Series(myarr)
ser_dict = pd.Series(mydict)

ser_list, ser_arr, ser_dict
```

## Bài 4.2

```
ser = pd.Series(mydict)
df = ser.reset_index()
df.columns = ['index', 'value']
df
```

## Bài 4.3

```
ser1 = pd.Series(list('abcdefghijklmnopqrstuvwxyz'))
ser2 = pd.Series(np.arange(26))
df = pd.concat([ser1, ser2], axis=1)
df
```

## Bài 4.4

```
ser1 = pd.Series([1,2,3,4,5])
ser2 = pd.Series([4,5,6,7,8])

# 1
ser1[~ser1.isin(ser2)]

# 2
pd.Series(np.setxor1d(ser1, ser2))
```

## Bài 4.5

```
ser = pd.Series(np.random.normal(10,5,25))
ser.min(), ser.quantile(0.25), ser.median(), ser.quantile(0.75), ser.max()
```

## Bài 4.6

```
ser = pd.Series(np.take(list('abcdefgh'), np.random.randint(8, size=30)))
ser.value_counts()
```

## Bài 4.7

```
np.random.seed(100)
ser = pd.Series(np.random.randint(1,5,12))
top2 = ser.value_counts().nlargest(2).index
ser.apply(lambda x: x if x in top2 else 'Other')
```

## Bài 4.8

```
ser = pd.Series(np.random.random(20))
pd.qcut(ser, 10, labels=[f'D{i}' for i in range(1,11)])
```
