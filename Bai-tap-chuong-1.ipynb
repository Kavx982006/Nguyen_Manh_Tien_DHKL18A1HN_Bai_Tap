### Task 1:

```python
class HinhChuNhat:
    def __init__(self):
        self.chieu_dai = 0
        self.chieu_rong = 0

    def nhap(self):
        self.chieu_dai = float(input("Nhập chiều dài: "))
        self.chieu_rong = float(input("Nhập chiều rộng: "))

    def chu_vi(self):
        return 2 * (self.chieu_dai + self.chieu_rong)

    def dien_tich(self):
        return self.chieu_dai * self.chieu_rong

    def in_thong_tin(self):
        print("Chiều dài:", self.chieu_dai)
        print("Chiều rộng:", self.chieu_rong)
        print("Chu vi:", self.chu_vi())
        print("Diện tích:", self.dien_tich())

hcn = HinhChuNhat()
hcn.nhap()
hcn.in_thong_tin()
```

### Task 2:

```python
class ThiSinh:
    def __init__(self):
        self.ho_ten = ""
        self.toan = 0
        self.ly = 0
        self.hoa = 0

    def nhap(self):
        self.ho_ten = input("Nhập họ tên: ")
        self.toan = float(input("Nhập điểm Toán: "))
        self.ly = float(input("Nhập điểm Lý: "))
        self.hoa = float(input("Nhập điểm Hóa: "))

    def tong_diem(self):
        return self.toan + self.ly + self.hoa

    def in_thong_tin(self):
        print(f"Họ tên: {self.ho_ten}")
        print(f"Toán: {self.toan}, Lý: {self.ly}, Hóa: {self.hoa}")
        print(f"Tổng điểm: {self.tong_diem()}")
        print("----------------------")


# ===== Chương trình chính =====
ds = []
n = int(input("Nhập số lượng thí sinh: "))

for i in range(n):
    print(f"\nNhập thí sinh thứ {i+1}:")
    ts = ThiSinh()
    ts.nhap()
    ds.append(ts)

# Sắp xếp theo tổng điểm giảm dần
ds.sort(key=lambda x: x.tong_diem(), reverse=True)

print("\nDANH SÁCH THÍ SINH TRÚNG TUYỂN (ĐIỂM CHUẨN = 20)")
for ts in ds:
    if ts.tong_diem() >= 20:
        ts.in_thong_tin()
```

### Task 3: 

```python
class PhanSo:
    def __init__(self):
        self.tu = 0
        self.mau = 1

    def hop_le(self):
        return self.mau != 0

    def nhap(self):
        self.tu = int(input("Nhập tử số: "))
        self.mau = int(input("Nhập mẫu số: "))
        while not self.hop_le():
            print("Mẫu số phải khác 0. Nhập lại!")
            self.mau = int(input("Nhập mẫu số: "))

    def in_phan_so(self):
        print(f"Phân số: {self.tu}/{self.mau}")


# ===== Chương trình chính =====
ps = PhanSo()
ps.nhap()
ps.in_phan_so()
```

### Task 4:

```python
class Stack:
    def __init__(self, size):
        self.size = size
        self.stack = []
    
    def isEmpty(self):
        return len(self.stack) == 0

    def isFull(self):
        return len(self.stack) == self.size

    def push(self, x):
        if self.isFull():
            print("Stack đã đầy!")
        else:
            self.stack.append(float(x))

    def pop(self):
        if self.isEmpty():
            print("Stack rỗng!")
            return None
        return self.stack.pop()


# ===== Chương trình chính =====
n = int(input("Nhập kích thước stack: "))
st = Stack(n)

while True:
    print("\n1. Push")
    print("2. Pop")
    print("3. Thoát")
    chon = int(input("Chọn: "))

    if chon == 1:
        x = float(input("Nhập giá trị: "))
        st.push(x)
    elif chon == 2:
        x = st.pop()
        if x is not None:
            print("Giá trị lấy ra:", x)
    else:
        break
```

```python
class Stack:
    def __init__(self, size):
        self.size = size
        self.stack = []

    def push(self, x):
        if len(self.stack) < self.size:
            self.stack.append(x)

    def pop(self):
        if not self.isEmpty():
            return self.stack.pop()

    def isEmpty(self):
        return len(self.stack) == 0

    def isFull(self):
        return len(self.stack) == self.size

    def count(self):
        return len(self.stack)
```

### Task 7:

```python
class Date:
    def __init__(self, d=1, m=1, y=2000):
        self.day = d
        self.month = m
        self.year = y

    def display(self):
        print(f"{self.day}/{self.month}/{self.year}")

    def next(self):
        self.day += 1
        if self.day > 30:
            self.day = 1
            self.month += 1
        if self.month > 12:
            self.month = 1
            self.year += 1
```

### Task 8:

```python
class Employee:
    def __init__(self, name, birth, join):
        self.name = name
        self.birth = birth
        self.join = join

    def display(self):
        print("Họ tên:", self.name)
        print("Ngày sinh:", end=" ")
        self.birth.display()
        print("Ngày vào công ty:", end=" ")
        self.join.display()


# Chương trình chính
b = Date(1, 1, 2004)
j = Date(1, 9, 2024)
e = Employee("Nguyen Van A", b, j)
e.display()
```

```python
class DaGiac:
    def __init__(self, a, b):
        self.a = a
        self.b = b

class BinhHanh(DaGiac):
    def chu_vi(self):
        return 2 * (self.a + self.b)

    def dien_tich(self, h):
        return self.a * h

class ChuNhat(BinhHanh):
    def dien_tich(self):
        return self.a * self.b

class Vuong(ChuNhat):
    def __init__(self, a):
        super().__init__(a, a)
```

```python
import math

class Diem:
    def __init__(self, x, y):
        self.x = x
        self.y = y

class Elip(Diem):
    def __init__(self, x, y, a, b):
        super().__init__(x, y)
        self.a = a
        self.b = b

    def dien_tich(self):
        return math.pi * self.a * self.b

class DuongTron(Elip):
    def __init__(self, x, y, r):
        super().__init__(x, y, r, r)
```
