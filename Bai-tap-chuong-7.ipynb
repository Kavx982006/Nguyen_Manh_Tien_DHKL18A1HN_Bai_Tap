## Bài 7.5 — Hiển thị ảnh từ file

```python

import tkinter as tk
from tkinter import messagebox
from PIL import Image, ImageTk

root = tk.Tk()
root.title("Bài 7.5 — Xem ảnh")

try:
    img = Image.open("Oto.png")   # cần file cùng thư mục
    photo = ImageTk.PhotoImage(img)
    lbl = tk.Label(root, image=photo)
    lbl.pack()
except Exception as e:
    messagebox.showerror("Lỗi", f"Không thể tải ảnh: {e}")

root.mainloop()
```

## Bài 7.6 — Nhập chuỗi và đảo ngược

```python

import tkinter as tk

root = tk.Tk()
root.title("Bài 7.6")

def reverse_text():
    t = entry.get()
    result.config(text=t[::-1])

tk.Label(root, text="Enter a word").pack()
entry = tk.Entry(root)
entry.pack()

tk.Button(root, text="Validate", command=reverse_text).pack(pady=5)
result = tk.Label(root, text="")
result.pack()

root.mainloop()
```

## Bài 7.7 — Nhập số N và tính tổng 1 + 2 + ... + N

```python

import tkinter as tk

root = tk.Tk()
root.title("Bài 7.7")

def calc_sum():
    try:
        n = int(entry.get())
        s = n*(n+1)//2
        result.config(text=f"Total = {s}")
    except:
        result.config(text="Input không hợp lệ")

tk.Label(root, text="Enter N").pack()
entry = tk.Entry(root)
entry.pack()

tk.Button(root, text="Validate", command=calc_sum).pack()
result = tk.Label(root, text="")
result.pack()

root.mainloop()
```

## Bài 7.8 — Hiển thị các ước của N

```python

import tkinter as tk

root = tk.Tk()
root.title("Bài 7.8")

def show_divisors():
    try:
        n = int(entry.get())
        ds = [str(i) for i in range(1, n+1) if n % i == 0]
        result.config(text=" , ".join(ds))
    except:
        result.config(text="Input lỗi")

tk.Label(root, text="Enter N").pack()
entry = tk.Entry(root)
entry.pack()
tk.Button(root, text="Validate", command=show_divisors).pack()
result = tk.Label(root, text="")
result.pack()

root.mainloop()
```

## Bài 7.9 — GCD & LCM (bội số chung, ước số chung lớn nhất)

```python

import tkinter as tk
import math

root = tk.Tk()
root.title("Bài 7.9")

def compute(func):
    try:
        a = int(a_entry.get())
        b = int(b_entry.get())
        if func == "GCD":
            res = math.gcd(a, b)
        else:
            res = abs(a*b)//math.gcd(a,b)
        result.config(text=str(res))
    except:
        result.config(text="Lỗi")

tk.Label(root, text="Enter value of m").grid(row=0, column=0)
a_entry = tk.Entry(root); a_entry.grid(row=0, column=1)

tk.Label(root, text="Enter value of n").grid(row=1, column=0)
b_entry = tk.Entry(root); b_entry.grid(row=1, column=1)

tk.Button(root, text="GCD", command=lambda: compute("GCD")).grid(row=2, column=0)
tk.Button(root, text="LCM", command=lambda: compute("LCM")).grid(row=2, column=1)

result = tk.Label(root, text="")
result.grid(row=3, column=0, columnspan=2)

root.mainloop()
```

## Bài 7.10 — Tính tiền taxi (theo bảng giá bài 7.14 tham khảo)

```python

import tkinter as tk

root = tk.Tk()
root.title("Bài 7.10 — Taxi")

def calc():
    try:
        d = float(entry.get())
        cost = 12000  # ví dụ 0.8 km đầu
        if d > 0.8:
            cost += (d-0.8)*15000
        result.config(text=f"Fare: {cost:,.0f} VND")
    except:
        result.config(text="Input lỗi")

tk.Label(root, text="Distance (km)").pack()
entry = tk.Entry(root); entry.pack()
tk.Button(root, text="Calculate", command=calc).pack()
result = tk.Label(root, text="")
result.pack()

root.mainloop()
```

## Bài 7.11 — Kiểm tra từ đối xứng

```python

import tkinter as tk

root = tk.Tk()
root.title("Bài 7.11")

def check():
    t = entry.get()
    result.config(text="Palindrome" if t == t[::-1] else "Not palindrome")

tk.Label(root, text="Enter word").pack()
entry = tk.Entry(root); entry.pack()
tk.Button(root, text="Validate", command=check).pack()
result = tk.Label(root, text=""); result.pack()
root.mainloop()
```

## Bài 7.12 — BMI Calculator

```python

import tkinter as tk

root = tk.Tk()
root.title("BMI")

def calc_bmi():
    try:
        w = float(weight.get())
        h = float(height.get())/100
        bmi = w/(h*h)
        result.config(text=f"BMI = {bmi:.2f}")
    except:
        result.config(text="Lỗi nhập")

tk.Label(root, text="Cân nặng (kg)").pack()
weight = tk.Entry(root); weight.pack()

tk.Label(root, text="Chiều cao (cm)").pack()
height = tk.Entry(root); height.pack()

tk.Button(root, text="Tính BMI", command=calc_bmi).pack()
result = tk.Label(root, text=""); result.pack()
root.mainloop()
```

## Bài 7.13 — Xây dựng ứng dụng nhập điểm và phân loại (giả lập)

```python

import tkinter as tk

root = tk.Tk()
root.title("Bài 7.13")

def classify():
    try:
        s = float(score.get())
        if s >= 8: r="Giỏi"
        elif s >= 6.5: r="Khá"
        elif s >= 5: r="TB"
        else: r="Yếu"
        result.config(text=r)
    except:
        result.config(text="Lỗi")

tk.Label(root, text="Nhập điểm").pack()
score = tk.Entry(root); score.pack()
tk.Button(root, text="Validate", command=classify).pack()
result = tk.Label(root, text=""); result.pack()
root.mainloop()
```
